```python
import pandas as pd
import numpy as np

DM='data/NCD_RisC_Lancet_2016_DM_age_standardised_countries.csv'
BMI='data/NCD_RisC_Lancet_2017_BMI_age_standardised_countries.csv'
BP='data/NCD_RisC_Lancet_2016_BP_age_standardised_countries.csv'

df_DM=pd.read_csv(DM,encoding='iso-8859-1')
df_BMI=pd.read_csv(BMI,encoding='iso-8859-1')
df_BP=pd.read_csv(BP,encoding='iso-8859-1')

```

```python
df_allData = pd.merge(df_DM,df_BMI,on=['Country/Region/World','ISO','Sex','Year'])
```

```python
df_allData = df_allData.merge(df_BP,on=['Country/Region/World','ISO','Sex','Year'])
```
